```python
from pathlib import Path
import json

path='./qa-datasets/in-domain-train/newsqa_train.json'

def make_data_form(path):
    path = Path(path)

    #'question': 'str' ,'context': 'str'
    with open(path,'rb') as f:
        dict = json.load(f)
        new_data_form = []
        
        for data in dict['data']:
            #instance 마다 하나씩 담아서 new_data_form으로 전달
            new_dict = {'answers':{'text':[],'answer_start':[]}}
            #all data
            for paragraphs in data['paragraphs']: 
                #context
                new_dict['context'] = paragraphs['context']
                #question
                new_dict['question'] = paragraphs['qas'][0]['question']
                #qas
                for qas in paragraphs['qas']:
                    #answers, id, is_impossible, question
                    #answer_start, text
                    for answers in qas['answers']:
                        new_dict['answers']['answer_start'].append(answers['answer_start'])
                        new_dict['answers']['text'].append(answers['text'])
            #하나의 instance info를 담았으면, new_data_form 바구니에 넣어주기
            new_data_form.append(new_dict)
            
    return new_data_form
    
new_data_form = make_data_form(path)
len(new_data_form)
```

```python
"""data form이 실제로 유효한지 확인"""
from transformers import AutoTokenizer

model_checkpoint = "distilbert-base-uncased" # distilbert-base-cased-distilled-squad
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
"""Tokenizer로 check"""
context = new_data_form[1]["context"]
question = new_data_form[1]["question"]

"""[CLS] question [SEP] context [SEP]로 변형(id 형태로)"""
inputs = tokenizer(question, context)
"""ID 형태를 다시 원래 text로 decoding"""
tokenizer.decode(inputs["input_ids"])
```
